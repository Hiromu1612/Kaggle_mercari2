```python
import matplotlib.pyplot as plt
import numpy as np

# データ
models = ['RFR', 'LightGBM', 'Ridge']
rmsle_values = [0.54197, 0.54516, 0.54257]
goal_value = 0.42275

# 棒グラフの色
colors = ['blue', 'orange', 'green']

# グラフの作成
plt.figure(figsize=(8, 5))
bars = plt.bar(models, rmsle_values, color=colors)

# 目標値の破線を追加
plt.axhline(y=goal_value, color='red', linestyle='--', label='Target Value (0.42275)')

# データラベルの追加
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 5), ha='center', va='bottom')

# ラベルとタイトル
plt.title('RMSLE Comparison of Different Models')
plt.xlabel('Models')
plt.ylabel('RMSLE')
plt.ylim(0.4, max(rmsle_values) + 0.1)

# 凡例の追加
plt.legend()

# グラフの表示
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データ
models = ['RFR', 'LightGBM', 'Ridge']
ver1_values = [0.54197, 0.54516, 0.54257]
ver2_values = [0.56080, 0.61183, 0.73832]
goal_value = 0.42275

# X軸の位置
x = np.arange(len(models))

# グラフの幅
width = 0.35

# グラフの作成
plt.figure(figsize=(10, 6))
bars1 = plt.bar(x - width/2, ver1_values, width, label='Version 1', color='blue')
bars2 = plt.bar(x + width/2, ver2_values, width, label='Version 2', color='orange')

# 目標値の破線を追加
plt.axhline(y=goal_value, color='red', linestyle='--', label='Target Value (0.42275)')

# データラベルの追加
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 5), ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 5), ha='center', va='bottom')

# ラベルとタイトル
plt.title('RMSLE Comparison Between Version 1 and Version 2')
plt.xlabel('Models')
plt.ylabel('RMSLE')
plt.xticks(x, models)
plt.ylim(0.4, max(ver2_values) + 0.1)

# 凡例の追加
plt.legend()

# グラフの表示
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データ
models = ['LightGBM', 'Ridge']
ver1_values = [0.54516, 0.54257]
ver2_values = [0.61183, 0.73832]
ver3_values = [0.52276, 0.46294]
goal_value = 0.42275

# X軸の位置
x = np.arange(len(models))

# グラフの幅
width = 0.25

# グラフの作成
plt.figure(figsize=(10, 6))
bars1 = plt.bar(x - width, ver1_values, width, label='Version 1', color='blue')
bars2 = plt.bar(x, ver2_values, width, label='Version 2', color='orange')
bars3 = plt.bar(x + width, ver3_values, width, label='Version 3', color='green')

# 目標値の破線を追加
plt.axhline(y=goal_value, color='red', linestyle='--', label='Target Value (0.42275)')

# データラベルの追加
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 5), ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 5), ha='center', va='bottom')

for bar in bars3:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 5), ha='center', va='bottom')

# ラベルとタイトル
plt.title('RMSLE Comparison From Version 1 to Version 3')
plt.xlabel('Models')
plt.ylabel('RMSLE')
plt.xticks(x, models)
plt.ylim(0.4, max(ver2_values) + 0.1)

# 凡例の追加
plt.legend()

# グラフの表示
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データ
versions = ['Ver.1', 'Ver.2', 'Ver.3', 'Ver.4', 'Target']
scores = [0.54197, 0.56080, 0.46816, 0.45222, 0.42275]
ranks = [1875, 1794, 1456, 1291, 476]
top_percentages = [79, 75, 61, 54, 20]

# X軸の位置
x = np.arange(len(versions))

# グラフの幅
width = 0.4

# グラフの作成
plt.figure(figsize=(10, 6))
bars = plt.bar(x, scores, width, color='skyblue')

# データラベルの追加
for i, bar in enumerate(bars):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 5), ha='center', va='bottom')
    plt.text(bar.get_x() + bar.get_width()/2, yval - 0.02, f'Rank: {ranks[i]}, Top: {top_percentages[i]}%', ha='center', va='top', fontsize=10)

# ラベルとタイトル
plt.title('Comparison of Versions and Target')
plt.xlabel('Versions')
plt.ylabel('Score (RMSLE)')
plt.xticks(x, versions)
plt.ylim(0.4, 0.575)

# グラフの表示
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
